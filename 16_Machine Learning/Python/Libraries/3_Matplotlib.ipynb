# Matplotlib

Matplotlib is a widely-used plotting library in Python for creating static, animated, and interactive visualizations. It is particularly useful for generating plots, histograms, bar charts, scatter plots, and more. Heres a detailed overview:
*********************************
__1. Overview:__

Matplotlib is part of the larger PyData stack, which includes libraries like NumPy, Pandas, and SciPy, making it well-suited for data visualization in scientific computing and data analysis.

__Key Features:__

Produces publication-quality figures in various formats (PNG, PDF, SVG, etc.).
Supports multiple types of plots: line, scatter, bar, histogram, 3D plots, and more.
Can be used in Python scripts, Python and IPython shells, Jupyter notebooks, and web application servers.
Highly customizable: control over every aspect of plot appearance (labels, ticks, colors, styles, etc.).

__Linear Graph__

```python
x = [1,2,3,4,5]
y = [5,6,7,8,9]

plt.plot(x,y)
plt.show
```

```python
x = [1,2,3,4,5]
y = [5,6,7,8,9]
c = 'r'
plt.plot(x,y,'r') #3rd varibale that mention the color
plt.show()
```

```python
x =[1,6,15,7,8]
y =[1,8,5,0,6]

plt.plot(x,y)
plt.show()
```

```python
# highlight the co-ordinate point
x =[1,6,15,7,8]
y =[1,8,5,0,6]

plt.plot(x,y, marker = 'o')
plt.show()

# marker : it denote the dot 
```

```python
# make x , y axis (alag-alag)
x =[1,6,15,7,8]
y =[1,8,5,0,6]

plt.plot(x,marker = 'o')
plt.plot(y, marker = '<')

plt.show()

# marker : it denote the dot 
```

# Bar - Graph

```python
# Vertical Bar graph

x =[1,2,3,4,5]
y =[6,7,8,9,10]

plt.bar(x,y)
plt.show()
```

```python
# Color change
x =[1,2,3,4,5]
y =[6,7,8,9,10]
col = ['y','r','g','purple','lightgreen']

plt.bar(x,y,color = col)
plt.show()
```

```python
import numpy as np
```

```python
x = np.array(['a','b','c','d','e'])
y = np.array([1,2,3,4,5])

plt.barh(x,y)
plt.show()

# Where: barh - h is denoted as horizontal
```

```python
x = np.array(['a','b','c','d','e'])
y = np.array([1,2,3,4,5])
# col = ['y','r','g','purple','lightgreen']
plt.xlabel("Number")
plt.ylabel("Month")
plt.barh(x,y,color = '#4CAF50',height=0.3)
plt.show()

# Where: barh - h is denoted as horizontal
```

# Scatter Plot

```python
import matplotlib.pyplot as plt
```

```python
x =[1,2,3,4,5]
y =[6,7,8,9,10]
col = ['y','r','g','purple','lightgreen']

plt.title("Scatter Plot")
plt.xlabel("Number")
plt.ylabel("Month")
plt.scatter(x,y,color = col)
plt.show()
```
