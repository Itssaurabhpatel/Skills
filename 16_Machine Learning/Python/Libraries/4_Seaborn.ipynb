# Seaborn

 Seaborn is a Python data visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. It simplifies the process of creating complex visualizations and integrates well with Pandas for data manipulation. Here's an overview of Seaborn:
************************
Key Features:

1. Built-in Themes and Color Palettes: Seaborn comes with aesthetically pleasing themes and color schemes.
1. Statistical Plots: Seaborn makes it easy to visualize distributions and relationships between data.
1. Integration with Pandas: Seamless interaction with Pandas DataFrames makes it a go-to for visualizing real-world datasets.
1. Dataset-Oriented API: Works efficiently with entire datasets using DataFrames and arrays for plotting.

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [4,7,5,8,9,6]
```

__Using Matplotlib__

```python
plt.plot(x,y)
```

__Using Seaborn__

```python
sns.lineplot(x=x,y=y)
plt.show()
```

```python
# It give additional information like - label

import pandas as pd

# Data
var = [1, 2, 3, 4, 5, 6]
var1 = [4, 7, 5, 8, 9, 6]

# Creating DataFrame
df = pd.DataFrame({
    'var': var,
    'var1': var1
})

# Plotting the lineplot using the DataFrame
sns.lineplot(x='var', y='var1', data=df)

# Display the plot
plt.show()
```

```python
df1 = sns.load_dataset('penguins')
```

```python
df1
```

```python
# Problems on datasets-making lineplot
sns.lineplot(x='bill_length_mm', y='flipper_length_mm',data=df1)
plt.show()
```

```python
# devide based on catogorical- male and female
sns.lineplot(x='bill_length_mm', y='flipper_length_mm',data=df1, hue='sex')
plt.show()

# Where :
'''
hue : When you specify the hue parameter, Seaborn automatically splits the data by the categories of the chosen variable and assigns different colors to each category.
'''
```

```python
sns.lineplot(x='bill_length_mm', y='flipper_length_mm', data=df1, hue='sex', palette='Greens')
```

```python
#marker
sns.lineplot(x='bill_length_mm', y='flipper_length_mm', data=df1, hue='sex', palette='Greens',style= 'sex', markers=['o','>'])
```

__Barplot in Seaborn__

```python
import pandas as pd
```

```python
df1 = sns.load_dataset('penguins')
```

```python
df1
```

```python
# Bar graph
df1['island']
```

```python
sns.barplot(x=df1.island, y=df1.bill_length_mm)
plt.show()
```

```python
# other way
sns.barplot(x='island', y='bill_length_mm', data=df1)
plt.show()
```

```python
# other way
sns.barplot(x='island', y='bill_length_mm', data=df1,)
plt.show()
```

```python
# categories
sns.barplot(x='island', y='bill_length_mm', data=df1, hue='sex')
plt.show()
```

```python
# orderizing

order1 =['Dream','Torgersen','Biscoe']
sns.barplot(x='island' , y='bill_length_mm' ,data=df1, hue='sex', order=order1)
plt.show()
```

```python
# change order of male and female based on individual bar
order1 =['Dream','Torgersen','Biscoe']
sns.barplot(x='island' , y='bill_length_mm' ,data=df1, hue='sex', order=order1, hue_order=['Female','Male'])
plt.show()
```

```python
# Saturation[0-1] -parameter : used for color  
order1 =['Dream','Torgersen','Biscoe']
sns.barplot(x='island' , y='bill_length_mm' ,data=df1, hue='sex', order=order1, hue_order=['Female','Male'], saturation=0.2)
plt.show()
```
